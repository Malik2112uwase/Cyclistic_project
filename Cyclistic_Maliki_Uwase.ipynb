Data manipulation

```python
#Transform
data_trips_df['started_date'] = pd. to_datetime(data_trips_df['started_date'])
data_trips_df['ended_date'] = pd. to_datetime(data_trips_df['ended_date'])
```

```python
#Loop to display number of casual riders and member riders every month 
months_data=[
    {
        'start': '2021-01-01',
        'end': '2021-01-31'
    },
    {
        'start': '2021-02-01',
        'end': '2021-02-28'
    },
    {
        'start': '2021-03-01',
        'end': '2021-03-31'
    }, 
    {
        'start': '2021-04-01',
        'end': '2021-04-30'
    },
    {
        'start': '2021-05-01',
        'end': '2021-05-31'
    },
    {
        'start': '2021-06-01',
        'end': '2021-01-30'
    },
    {
        'start': '2021-07-01',
        'end': '2021-07-31'
    },
    {
        'start': '2021-08-01',
        'end': '2021-08-31'
    },
    {
        'start': '2021-09-01',
        'end': '2021-09-30'
    },
    {
        'start': '2021-10-01',
        'end': '2021-10-31'
    },
    {
        'start': '2021-11-01',
        'end': '2021-11-30'
    },
    {
        'start': '2021-12-01',
        'end': '2021-12-31'
    }
]

all_months_data=[]
casual_each_month=[]
member_each_month=[]
for month_d in months_data:
    
    Just_one_data=data_trips_df[(data_trips_df['started_date'] >=pd. to_datetime(month_d['start'])) & (data_trips_df['ended_date']<=pd. to_datetime(month_d['end']))]
    all_months_data.append(Just_one_data)
    casual_each_month.append(Just_one_data[Just_one_data['casual'] == 1])
    member_each_month.append(Just_one_data[Just_one_data['member'] == 1])
    
    
casuals= [len(i) for i in casual_each_month]
members= [len(i) for i in member_each_month]

```
